```python
import csv

# 读取 TSV 文件
def read_tsv(file_path):
    data = {}
    with open(file_path, 'r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter='\t')
        for row in reader:
            if len(row) >= 3:
                gene, go_id, _ = row[:3]
                if gene not in data:
                    data[gene] = set()
                data[gene].add(go_id)
    return data

# 合并多个文件的数据
def merge_files(file_paths):
    merged_data = {}
    for file_path in file_paths:
        data = read_tsv(file_path)
        for gene, go_ids in data.items():
            if gene not in merged_data:
                merged_data[gene] = set()
            merged_data[gene].update(go_ids)
    return merged_data

# 文件路径
file_paths = [
    './deepgo2-main/output/found_genes_preds_bp.tsv',
    './deepgo2-main/output/found_genes2_preds_bp.tsv',
    './deepgo2-main/output/found_genes3_preds_bp.tsv'
]

# 合并文件数据
merged_data = merge_files(file_paths)

# 输出结果
print(merged_data)
```
