### pearson_delta_de

```python
import os
import pandas as pd
import pickle

# Define names and methods
names = ['AdamsonWeissman2016_GSM2406675_1', 'AdamsonWeissman2016_GSM2406677_2', 'AdamsonWeissman2016_GSM2406681_3', 
         'DatlingerBock2017_stimulated', 'DatlingerBock2017_unstimulated', 'DatlingerBock2021_stimulated', 
         'DatlingerBock2021_unstimulated', 'Dixit_combined', 'Dixit_GSM2396858', 'Dixit_GSM2396861', 
         'NormanWeissman2019', 'PapalexiSatija2021_eccite_arrayed_RNA', 'PapalexiSatija2021_eccite_RNA', 
         'Replogle_k562_essential', 'Replogle_rpe1_essential', 'ShifrutMarson2018', 'TianKampmann2019_day7neuron', 
         'TianKampmann2019_iPSC', 'TianKampmann2021_CRISPRa', 'TianKampmann2021_CRISPRi', 'XuCao2023']
methods = ['no_perturb', 'grn', 'cpa', 'gears', 'scgpt', 'scfoundation', 'scELMo_GPT35', 'new_geneformer', 'new_scbert', 'new_GenePT', 'gears_esm', 'attnpert', 'sclong']

# Initialize an empty list to store results
results = []

# Loop over each dataset name, method, and split
for name in names:
    for method in methods:
        base_path = f'/home/share/huadjyin/home/zhoumin3/zhoumin/model_benchmark/01_A_results/{name}/{method}'
        for i in range(1, 6):  # Assuming splits are 1 to 5
            pkl_path = f'{base_path}/split{i}/{name}_split{i}_deeper_res.pkl'

            # Check if the file exists
            if os.path.exists(pkl_path):
                try:
                    # Load the pickle file
                    with open(pkl_path, 'rb') as f:
                        pert_res = pickle.load(f)

                    # Extract data for each perturbed gene
                    for gene, metrics in pert_res.items():
                        if 'pearson_delta_de' in metrics:
                            results.append({
                                'Dataset': name,
                                'Method': method,
                                'Split': i,
                                'Perturbed Gene': gene,
                                'pearson_delta_de': metrics['pearson_delta_de']
                            })
                except Exception as e:
                    print(f"Error reading {pkl_path}: {e}")

# Convert results to a DataFrame
df_results = pd.DataFrame(results)
```

```python
import pickle
import os
import pandas as pd

results = []
columns = ['subgroup', 'method', 'split', 'Perturbed Gene', 'pearson_delta_de']
methods = ['no_perturb', 'grn', 'cpa', 'gears', 'scgpt', 'scfoundation', 'scELMo_GPT35', 
           'new_geneformer', 'new_scbert', 'new_GenePT', 'gears_esm', 'attnpert', 'sclong']

for method in methods:
    base_path = f'/home/share/huadjyin/home/zhoumin3/zhoumin/model_benchmark/01_A_results/NormanWeissman2019/{method}'
    
    # Iterate over split1 to split5
    for i in range(1, 6):
        # Load subgroup file
        subgroup_path = f'/home/share/huadjyin/home/zhoumin3/zhoumin/Gears_change/geneformer/All_data/normanweissman2019/splits/normanweissman2019_simulation_{i}_0.75_subgroup.pkl'
        if os.path.exists(subgroup_path):
            subgroup = pickle.load(open(subgroup_path, "rb"))
        else:
            print(f"File not found: {subgroup_path}")
            continue
        
        # Load test_pert_res file
        test_pert_res_path = f'{base_path}/split{i}/NormanWeissman2019_split{i}_deeper_res.pkl'
        if os.path.exists(test_pert_res_path):
            with open(test_pert_res_path, 'rb') as f:
                test_pert_res = pickle.load(f)
        else:
            print(f"File not found: {test_pert_res_path}")
            continue
        
        # Loop over each subgroup in the current split
        for subgroup_name, pert_list in subgroup['test_subgroup'].items():
            for pert in pert_list:
                if pert in test_pert_res:
                    pert_res = test_pert_res[pert]
                    pearson_delta_de = pert_res.get('pearson_delta_de', None)  
                    
                    if pearson_delta_de is not None:
                        results.append({
                            'subgroup': subgroup_name,
                            'method': method,
                            'split': f'split{i}',
                            'Perturbed Gene': pert,
                            'pearson_delta_de': pearson_delta_de
                        })

df_results = pd.DataFrame(results, columns=columns)
df_results.to_csv('norman_subgroup_pearson_delta_de.csv', index=False)
df_results.head()
```
