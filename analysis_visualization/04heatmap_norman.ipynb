### combo_seen1

```python
filtered_df = df[df['subgroup'] == 'combo_seen1']
filtered_df['Unique_Perturbed_Gene'] = filtered_df['Perturbed Gene'] + '_' + filtered_df['split'].astype(str)
sampled_genes = filtered_df['Unique_Perturbed_Gene'].sample(n=40, random_state=42)
filtered_df = filtered_df[filtered_df['Unique_Perturbed_Gene'].isin(sampled_genes)]
pivot_df = filtered_df.pivot(index='Unique_Perturbed_Gene', columns='method', values='pearson_delta_de')
long_df = pd.melt(pivot_df.reset_index(), 
                  id_vars='Unique_Perturbed_Gene',  
                  var_name='method',
                  value_name='pearson_delta_de')
if long_df['pearson_delta_de'].dtype == 'object':
    long_df['pearson_delta_de'] = pd.to_numeric(long_df['pearson_delta_de'], errors='coerce')

long_df.dropna(subset=['pearson_delta_de'], inplace=True) 

plt.figure(figsize=(9,3.5))
cm = DotClustermapPlotter(data=long_df, x='Unique_Perturbed_Gene', y='method', value='pearson_delta_de',
                          c='pearson_delta_de', cmap='custom4', vmax=1, vmin=-1, s=1, marker='s', spines=True,
                          show_rownames=True, show_colnames=True, row_dendrogram=True, grid=None, 
                          col_dendrogram=True, row_dendrogram_size=8, col_dendrogram_size=8)

plt.show()
```
